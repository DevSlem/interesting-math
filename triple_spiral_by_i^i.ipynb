# Triple Spiral by $i^i$

### Sequential powers of imaginary number $i$ which is $i, i^i, i^{i^i}, \dots$

```python
import numpy as np

def make_sequential_pow_of_i(count: int) -> np.ndarray:
    """Returns i, i^i, i^i^i, ..."""
    data = [0] * count
    data[0] = 1j

    for i in range(count - 1):
        data[i + 1] = 1j ** data[i]
    
    data = np.array(data)
    return data
```

### Plot function

```python
import matplotlib.pyplot as plt

def plot_sequential_pow_of_i(count: int, marker: str = "-o", color="g"):
    """Plot points of sequential powers of imaginary number"""
    data = make_sequential_pow_of_i(count)
    plt.plot(data.real, data.imag, marker, c=color)
    plt.xlabel("real")
    plt.ylabel("imaginary")
    plt.show()
```

### Plot four points

You may not find any pattern of it with just 4 points.

```python
plot_sequential_pow_of_i(4)
```

### Plot ten points

Can you find any pattern? Points are reaching at the center asymptotically.

```python
plot_sequential_pow_of_i(10)
```

### Plot a lot of points

Okay. We will now plot a lot of points without lines. You can see the triple spiral.

```python
plot_sequential_pow_of_i(200, marker="o")
```
